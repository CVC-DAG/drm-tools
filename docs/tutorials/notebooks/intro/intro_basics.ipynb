# Getting Started with CVCDOCDB

Minimal end-to-end graph workflow with `NetworkXGraph` and `WeakNode`.


```python
import importlib.util

# Comprovació de presència del paquet
package_to_check = 'cvcdocdb'
spec = importlib.util.find_spec(package_to_check)

if spec is None:
    print(f'⚠️ {package_to_check} no està instal·lat. Iniciant instal·lació...')
    %pip install -q --upgrade cvcdocdb-tools
    print("✅ Instal·lació completada. L'estat del kernel PODRIA requerir un reinici.")
else:
    print(f'✅ {package_to_check} ja està present al sistema. Saltant instal·lació.')
```

```python
try:
	from cvcdocdb import NetworkXGraph, Node, WeakNode
except ImportError:
	# Fallback for installs where WeakNode is not re-exported from cvcdocdb.__init__
	from cvcdocdb import NetworkXGraph
	from cvcdocdb.base import Node, WeakNode
graph = NetworkXGraph()
doc = Node(pk={"doc": "DOC-001"}, main_label="Document")
section = WeakNode(parent=doc, pk={"section": 1}, main_label="Section")
graph.insertNode(doc, replace=True)
graph.insertNode(section, insert_parent=True)
print(graph.get_nodes())
graph.close()
```
