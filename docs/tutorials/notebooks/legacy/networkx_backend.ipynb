```python
%pip install -q --upgrade git+https://github.com/CVC-DAG/drm-tools.git
import importlib.util
import subprocess
import sys
from pathlib import Path

if importlib.util.find_spec("drm.exemples") is None:
    repo_root = Path.cwd().resolve()
    while repo_root != repo_root.parent and not (repo_root / "setup.py").exists():
        repo_root = repo_root.parent
    if (repo_root / "setup.py").exists():
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", "-e", str(repo_root)])
    else:
        raise ModuleNotFoundError("drm.exemples not found. Install from a local clone with `pip install -e <repo_root>`.")
```

```python
from drm import NetworkXGraph
from drm.exemples import (
	load_bibliografia_openalex,
	load_got_characters,
	load_karate_club,
	load_movies_sample,
)
```

```python
graph = NetworkXGraph()

karate_stats = load_karate_club(graph)
print("Karate loaded:", karate_stats)

print("sample KarateMember nodes:", graph.find_nodes_by_property("source", "networkx.karate_club_graph")[:5])
```

```python
# OpenAlex: downloads bibliographic works and citation/authorship links.
biblio_stats = load_bibliografia_openalex(
	graph,
	query="graph database",
	per_page=15,
)
print("Bibliography loaded:", biblio_stats)

print("sample Paper nodes:", graph.find_nodes_by_property("source", "openalex")[:5])
```

```python
# Movies: works in NetworkX too, even though the source is used by the Neo4j tutorial.
movies_stats = load_movies_sample(graph, limit=10)
print("Movies loaded:", movies_stats)
```

```python
# Game of Thrones characters: same loader can populate either backend.
got_stats = load_got_characters(graph, limit=10)
print("GoT loaded:", got_stats)
```
